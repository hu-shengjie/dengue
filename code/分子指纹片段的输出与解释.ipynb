```python
import matplotlib.pyplot as plt
import math
from rdkit.Chem import Draw
%matplotlib inline

# 假设你已经定义了 mol 和 bi
bits = [1]
imgs = []

for bit in bits:
    mfp2_svg = Draw.DrawMorganBit(mol, bit, bi)
    imgs.append(mfp2_svg)

def displayimgsinrow(imgs, col=4, save_path=None):
    # 增加 figsize 和 dpi 以提高图像质量
    plt.figure(figsize=(30, 15))  # 增大图像尺寸
    columns = col
    rows = math.ceil(len(imgs) / columns)
    for i, image in enumerate(imgs):
        ax = plt.subplot(rows, columns, i + 1)
        ax.set_axis_off()
        plt.imshow(image)
    
    if save_path:
        # 增加 dpi 以提高分辨率
        plt.savefig(save_path, bbox_inches='tight', dpi=600)  # 使用更高的 dpi (600)
    plt.show()

# 调用函数显示图片并保存
displayimgsinrow(imgs, save_path='output_image_high_res.png')
```
